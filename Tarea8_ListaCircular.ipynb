```
class Nodo():
 
    def __init__(self,date):
        self.date=date
        self.next=None
 
class circularList():
 
    def __init__(self,node=None):
        self.__head=node
        if node:
            node.next=self.__head
 
    def is_empty(self):
        return self.__head == None
 
    def length(self):
        if self.is_empty():
            return 0
        count=1
        cur=self.__head
        while cur.next!=self.__head:
            count+=1
            cur=cur.next
        return count
 
    def insert(self,num):
        node=Nodo(num)
        if self.is_empty():
            self.__head=node
            node.next=node
        else:
            if node.date <= self.__head.date:
                cur=self.__head
                while cur.next!=self.__head:
                    cur=cur.next
                node.next=self.__head
                self.__head=node
                cur.next=self.__head
            else:
                cur=self.__head
                while cur.next!=self.__head:
                    cur=cur.next
                cur.next=node   
                node.next=self.__head
 
    def transversal (self):
        if self.is_empty():
            print ("----LISTA VACIA----")
            return
        
        cur = self.__head
        while cur.next != self.__head:
            print(cur.date,"--> ",end="")
            
            cur = cur.next
        print(cur.date,end="\n") 
 
    def search (self, value):
        counter = 0
        if self.is_empty():
            print ("----NO HAY NODO QUE MOSTRAR----")
            return
        cur = self.__head
        while cur.next != None:
            counter += 1
            if cur.date == value:
                print (f"'{value}' en la posición {counter}")
                return 
            elif counter == self.length():
                print (f"----VALOR '{value}' NO ENCONTRADO----")
                return
            else:
                cur = cur.next
 
    def remove(self, value):
 
        if self.is_empty():
            print ("----LISTA VACIA----")
            return
        head = self.__head
        curr = self.__head
        prev = curr.next
        if curr.date == value:
            self.delete_first(value)
            return
        while prev.next != head:
            prev = prev.next
 
        if curr.date == value:
            prev.next = curr.next
 
        curr = curr.next
        prev = prev.next
        while curr.date != value and curr != head:
            curr = curr.next
            prev = prev.next
 
        if curr.date == value:
            prev.next = curr.next
            
    def delete_first(self, value):
        p = self.__head
        if p is None:
            return None
        if p.date == value and p.next == p:
            return None
        while p.date != value:
            p = p.next
            if p.next is p and p.date != value:
                return head
        p.date = p.next.date
        if p.next == p:
            self.__head = p
        p.next = p.next.next
        return p
 
a = circularList()
print ("RESULTADOS DE LISTA VACIA")
print (f"¿Esta vacía? {a.is_empty()}")
a.search(8)
a.transversal()
a.remove(6)
print (f"Nodos en la lista: {a.length()}")
print ("\nRESULTADOS DE LISTA CON NODOS")
a.insert(2)
a.insert(3)
a.insert(1)
a.insert(4)
a.insert(0)
a.insert(5)
a.transversal()
print (f"Nodos en la lista: {a.length()}")
a.search(5)
a.search(0)
a.search('A')
a.remove(0)
a.remove(5)
a.transversal()
print (f"¿Esta vacía? {a.is_empty()}")
print (f"Nodos en la lista: {a.length()}")
```
