```
class NodoDoble:
    def __init__(self,value,siguiente=None,anterior=None):
        self.data = value
        self.next = anterior
        self.prev = siguiente
    
    def getNext(self):
        return self.next
 
    def getPrevious(self):
        return self.prev
 
    def setNext(self, newNext):
        self.next = newNext
 
    def setPrev(self, newPrevious):
        self.prev = newPrevious
 
    def __str__(self):
        return str(self.data)
 
    def printNode(self):
        print ( "Anterior:", self.prev,'|',"Valor:", self.data,"|", "Siguiente:", self.next)
 
class DoublyLinkedList:
    def __init__(self): 
        self.__head = None
        self.__tail = None
        self.__size = 0
        print(f"Head : {self.__head}")
        print(f"Tail : {self.__tail}")
 
    def get_size(self):
        return self.__size
   
    def is_empty(self):
        return self.__head == None
 
    def append(self, data):
        newNode = NodoDoble(data)
        current = self.__head
        if self.__tail == None: #Para una lista vacia
            self.__head = NodoDoble(data)
            self.__tail = NodoDoble(data)
            NodoDoble(data).prev = None
            NodoDoble(data).next = None
            self.__size+=1
        else:
            while current.getNext() != None:
                current = current.getNext()
            current.setNext(newNode)
            newNode.setPrev(current)
            self.__tail = newNode
            self.__size+=1
 
    def find_from_tail(self, value):
        contador = 0
        dato = None
        current = self.__tail
        if current == None: #Para una lista vacia
            print ("Lista vacia, no hay nodo que encontrar") 
            print ()
        else: 
            while current.data != None:
                contador+=1
                if current.data == value:
                    print ("------Buscando por tail------")
                    print (f"Dato {value} en la posición {contador}")
                    print ()
                    return
                current = current.prev
 
    def find_from_head(self, value):
        contador = 0
        dato = None
        current = self.__head
        if current == None: #Para una lista vacia
            print ("Lista vacia, no hay nodos que encontrar")
            print()
        else: 
            while current.data != None:
                contador+=1
                if current.data == value: 
                    print ("------Buscando por head------")
                    print (f"Dato {value} en la posición {contador}") 
                    print ()
                    return
                current = current.next 
      
    def delete_nodo(self,value):     
        if self.__head == None: #Para una lista vacia
            print ("No hay nodos que eliminar")
        elif self.__head.data == value: 
            self.__head = self.__head.next
            self.__head.prev = None
 
        else: 
            current = self.__head
            while current.next != None and current.data != value:
                current = current.next
                if current.data == value:
                    current.prev.next = current.next
                    if current.next != None:
                        current.next.prev = current.prev
                    
 
    def remove_from_tail(self,value):
        contador = 0
        current = self.__tail
        if current == None: #Para una lista vacia
            print ("Lista vacia")
        elif value > self.__size: #Para un número mayor que el de nodos
            print ("-----Fuera de rango-----")
            print ()
            return
        else: 
            print("-------Eliminación por tail-------")
            while current.data != None:
                contador+=1
                if contador == value:
                    print (f"Valor {value}, se elimina nodo {current}")
                    self.delete_nodo(current.data)
                    print()
                    return
                current = current.prev 
        
    def remove_from_head(self,value):
        contador = 0
        current = self.__head
        if current == None:
            print ("Lista vacia") #Para una lista vacia
        elif value > self.__size:
            print ("-----Fuera de rango-----") #Para un número mayor que el de nodos
            print ()
            return
        else: 
            print("-------Eliminación por head-------")
            while current.data != None:
                contador+=1
                if contador == value:
                    print (f"Valor {value}, se elimina nodo {current}")
                    self.delete_nodo(current.data)
                    print()
                    return
                current = current.next      
 
    def transversal(self):
        current = self.__head
        if self.__head == None:
            print ("----Lista vacia, no hay nodos para mostrar----")
        else: 
            print ("----Recorrido empezando por head----")
            while current.next != None:
                current.printNode()
                current = current.next
            current.printNode()
        print ()
 
    def reverse_transversal(self): 
        current = self.__tail
        if self.__tail == None:
            print ("Lista vacia")
        else: 
            print ("----Recorrido empezando por tail----")
            while current.prev != None:
                current.printNode()
                current = current.prev
            current.printNode()
        print ()
 
 
l = DoublyLinkedList()
print(f"El tamaño de la lista es {l.get_size()}")
print(f"Lista vacia? {l.is_empty()} ")
l.find_from_head(52)
l.transversal()
l.append(52)
l.append(55)
l.append(88)
l.append(56)
l.append(65)
print(f"El tamaño de la lista es {l.get_size()}")
print(f"Lista vacia? {l.is_empty()}")
l.transversal()
l.reverse_transversal()
l.find_from_head(65)
l.find_from_tail(65)
#l.delete_nodo(56)
l.transversal()
l.remove_from_head(1)
l.transversal()
l.remove_from_tail(1)
l.transversal()
```
