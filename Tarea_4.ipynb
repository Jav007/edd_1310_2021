```
class Empleado:
    def __init__(self,numero=0,nombre="",appellidoPaterno="",apellidoMaterno="",horasExtra=0,sueldoBase=0,yearIngreso=0):
        self.numero=numero
        self.nombre=nombre
        self.apellidoPaterno=appellidoPaterno
        self.apellidoMaterno=apellidoMaterno
        self.horasExtra=horasExtra
        self.sueldoBase=sueldoBase
        self.yearIngreso=yearIngreso
        self.pagoC=0
        self.pagoEx=0
 
    def obtenerNombre(self):
        self.nombre
 
    def pagoExtra(self):
        if self.horasExtra!=0:
            if self.horasExtra > 0:
                self.pagoEx = self.horasExtra * 276.5
                return "Su pago extra correspondiente es $ "+ str(self.pagoEx)
        else:
            self.pagoEx = 0
            return "No cuentas con horas extras"
 
    def pagoComi(self):
        anio = 2020 - self.yearIngreso
        if anio!=0:
            comis = anio * 0.3
            self.pagoC = (self.sueldoBase*comis)
            return "Su pago de comision es $ "+ str(self.pagoC) + " por "+str(anio)+ " años en la empresa"
        else:
            anio=0
            return "Aun no cuentas con años de antiguedad"
 
    def __str__(self):
        cadenaPrint = "Trabajador :\n"+self.apellidoPaterno+" "+self.nombre+" ("+str(self.numero)+") "
        return cadenaPrint
 
    def pagoFinal(self):
        pagof = self.sueldoBase+self.pagoC + self.pagoEx
        return "Su pago final es: $"+str(pagof)
```

```
 
minimo=2020
maximo=0
tMayor=[]
tMenor=""
campo=[]
text = open("junio.dat",'r', encoding='latin1')
for line in text:
 
    if (line.count(' ')==0):
 
        campo = line.split(',')
        numeros = campo[0]
        nombres = campo[1]
        apellPate = campo[2]
        apellMate = campo[3]
        hrExt = int(campo[4])
        suelBase = int(campo[5])
        yearIngr = int(campo[6])
        if minimo <= yearIngr:
            tMenor = Empleado(numeros,nombres,apellPate,apellMate,hrExt,suelBase,yearIngr)
            minimo = yearIngr
 
        if minimo > yearIngr:
            tMayor = Empleado(numeros,nombres,apellPate,apellMate,hrExt,suelBase,yearIngr)
            maximo = yearIngr
 
        trab = Empleado(numeros,nombres,apellPate,apellMate,hrExt,suelBase,yearIngr)
        
        print(trab)
        print(trab.pagoExtra())
        print(trab.pagoComi())
        print(trab.pagoFinal())
        
print(F"\nEl empleado con menor antiguedad ingreso el {minimo} ")
print(f"Sus datos correspondientes son \n{tMenor}")
print(tMenor.pagoExtra())
print(tMenor.pagoComi())
print(tMenor.pagoFinal())
print(F"\nEl empleado con menor antiguedad ingreso el {maximo}")
print(f"Sus datos correspondientes son \n{tMayor}")
print(tMayor.pagoExtra())
print(tMayor.pagoComi())
print(tMayor.pagoFinal())
```
